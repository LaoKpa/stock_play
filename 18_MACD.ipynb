```python
def get_close_price(df):
    """ Get the simple moving average of a stock that's in a data frame. """
    df = df.sort_index(axis=0) ## We want the dates in ascending order
    close = np.zeros(len(df))

    for i in range(len(df)):
        close[i] = df['close'][i]

    return close
```

```python
def get_sma(df, period_length):
    """ Get the simple moving average of a stock that's in a data frame. """
    #df = df.sort_index(axis=0) ## We want the dates in ascending order
    sma = np.zeros(len(df))
    sum_period = np.zeros(len(df))
    count = 0

    for i in range(len(df)):
        count = 0
        for j in range(period_length):
            if (i - j) >= 0:
                sum_period[i] += df[i - j]
                count += 1
        sma[i] = sum_period[i] / count
    return sma
```

```python
close = get_close_price(df)
sma_20 = get_sma(df['close'], 20)
sma_50 = get_sma(df['close'], 50)
sma_100 = get_sma(df['close'], 100)
```

```python
g = get_sma(df['close'], 4)
assert(int(g[0]) == int(df['close'][0]))
assert(int(g[1]) == int((df['close'][0] + df['close'][1]) / 2))
assert(int(g[2]) == int((df['close'][0] + df['close'][1] + df['close'][2]) / 3))
assert(int(g[3]) == int((df['close'][0] + df['close'][1] + df['close'][2] + df['close'][3]) / 4))
assert(int(g[4]) == int((df['close'][1] + df['close'][2] + df['close'][3] + df['close'][4]) / 4))
```

```python
plt.figure(figsize=(16,6))
plt.plot(close, label='Stock Price')
plt.plot(sma_20, label='20 day SMA')
plt.plot(sma_50, label='50 day SMA')
plt.plot(sma_100, label='100 day SMA')
plt.legend()
plt.show()
```

```python
def get_ema(df, period_length, alpha):
    """ Get the explonential moving average of a stock that's in a data frame. """
    #df = df.sort_index(axis=0) ## We want the dates in ascending order
    ema = np.zeros(len(df))
    sum_period = np.zeros(len(df))
    count = 0
    if not ( 0 <= alpha <= 1):
        raise ValueError('Alpha should be between 0 and 1')

    for i in range(len(df)):
        count = 0
        this_alpha = alpha
        for j in range(period_length):
            if (i - j) >= 0:
                sum_period[i] += df[i - j] * this_alpha
                count += this_alpha
                #print(this_alpha / alpha)
                this_alpha *= alpha
        ema[i] = sum_period[i] / count
    return ema
```

## MACD

```python
alpha = 0.95
ema_12 = get_ema(close, 12, alpha)
ema_26 = get_ema(close, 26, alpha)
macd = ema_12 - ema_26
macd_ema9 = get_ema(macd, 9, alpha)
macd_diff = macd - macd_ema9

zero_line = np.zeros(len(macd))
```

```python
crossover_points = []
for i in range(len(macd)):
    if np.absolute(ema_12[i] - ema_26[i]) < 0.1:
        crossover_points.append((i, ema_12[i]))
```

```python
## Buy signals
buy_signals = []
sell_signals = []

for i in range(len(macd_diff) - 1):
    if macd_diff[i] < 0 and macd_diff[i+1] > 0:
        buy_signals.append((i, close[i]))
    elif macd_diff[i] > 0 and macd_diff[i+1] < 0:
        sell_signals.append((i, close[i]))
```

```python
plt.figure(figsize=(16,6))
#plt.scatter(crossover_points[:][0], crossover_points[:][1])
for i in range(len(crossover_points)):
    plt.scatter(crossover_points[i][0], crossover_points[i][1])
#plt.plot(crossover_points)
plt.show()
```

```python
viewing_window = 220

## EMA
plt.figure(figsize=(16,6))
plt.plot(close[-viewing_window:], label='Stock Price')
plt.plot(ema_12[-viewing_window:], label='12 day EMA')
plt.plot(ema_26[-viewing_window:], label='26 day EMA')
#plt.plot(crossover[-200:], label='C')
plt.title('EMA')

## EMA 12, 26 Crossover Lines
#for i in range(len(crossover_points)):
#    if crossover_points[i][0] > len(ema_12) - viewing_window:
#        plt.scatter(crossover_points[i][0] - (len(ema_12) - viewing_window), crossover_points[i][1], c='black')
        #print('ok')

## Buy and Sell Signals
for i in range(len(buy_signals)):
    if buy_signals[i][0] > len(ema_12) - viewing_window:
        plt.scatter(buy_signals[i][0] - (len(ema_12) - viewing_window), buy_signals[i][1], c='green')
for i in range(len(sell_signals)):
    if sell_signals[i][0] > len(ema_12) - viewing_window:
        plt.scatter(sell_signals[i][0] - (len(ema_12) - viewing_window), sell_signals[i][1], c='red')
        
plt.legend()

## MACD
plt.figure(figsize=(16,6))
plt.plot(zero_line[-viewing_window:], c='black')
plt.plot(macd[-viewing_window:], label='MACD', c='blue')
plt.plot(macd_ema9[-viewing_window:], label='MACD Signal Line', c='green')
plt.plot(macd_diff[-viewing_window:], label='MACD Difference', c='red', linestyle='dashed')
plt.title('MACD')
plt.legend()
plt.show()
```
