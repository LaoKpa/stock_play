## Goal of this notebook
 - Plot KDE (kernel density estimate) of distributions after each two day movement category combination
 - Take a look at the mean, variance, and level of normality of these curves

```python
from stock_utils import *
```

```python
df = pd.DataFrame()
df = df.from_csv('stock_data/spy.csv')
daily_movements = get_price_movement_percentages(df)
movement_categories = categorize_movements(daily_movements)
```

```python
three_day_trends_linear = get_trends_linear(movement_categories, daily_movements, 3)
```

```python
all_cats = ['bd', 'sd', 'sg', 'bg']
movements_after_category = []
for cat1 in all_cats:
    for cat2 in all_cats:
        movements_after_category.append(get_movements_after_trend(cat1 + '_' + cat2, three_day_trends_linear))
```

```python
## Big drop on the first day

plt.figure(figsize=(16,8))
plt.subplot(2, 2, 1)
x = movements_after_category[0]
n = len(movements_after_category[0])
plot_gaussian_categorical(x, title="Distribution of Movements after 2 Big Drops", n_cats=4, n_data=n)

plt.subplot(2, 2, 2)
x = movements_after_category[1]
n = len(movements_after_category[1])
plot_gaussian_categorical(x, title="Distribution of Movements after a Big Drop, then a Small Drop", n_cats=4, n_data=n)

plt.subplot(2, 2, 3)
x = movements_after_category[2]
n = len(movements_after_category[2])
plot_gaussian_categorical(x, title="Distribution of Movements after a Big Drop, then a Small Gain", n_cats=4, n_data=n)

plt.subplot(2, 2, 4)
x = movements_after_category[3]
n = len(movements_after_category[3])
plot_gaussian_categorical(x, title="Distribution of Movements after a Big Drop, then a Big Gain", n_cats=4, n_data=n)

plt.tight_layout()
plt.show()
```
