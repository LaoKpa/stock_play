(Note: we will be using the daily mid price to define a stock price)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
```

```python
df = pd.DataFrame()
df = df.from_csv('tsla.csv')
df['Mid'] = (df['High'] + df['Low']) / 2
```

```python
df.describe()
```

### 1)  Get the daily movement percentages of TSLA
(Note: For simplicity, we are assuming that the movement on day 1 is zero)

```python
movement = np.zeros(250)
last_price = -1
i = 0
for index, row in df.iterrows():
    if (last_price < 0):
        movement[i] = 0
    else:
        movement[i] = (100 * (row['Mid'] - last_price)) / last_price
    
    last_price = row['Mid']
    i += 1
```

### 2) Plot Stock Movement Percentages

```python
df['Movement'] = movement
```

```python
plt.figure()
df['Movement'].plot()
plt.show()
```

### 3) Plot the normal curve for this stock

```python
plt.plot()
df['Movement'].plot(kind='density')
plt.show()
```
